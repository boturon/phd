__log-linear detrending__

$$ y_t \equiv \log{y_t} $$
$$ y_t = a + bt + \varepsilon_t$$
$$ \text{secular trend: } \quad y_t^s = a + bt$$
$$ \text{cycle: } \quad y_t^c = \varepsilon_t$$

```python
X = [i for i in range(0, len(yh))]
X = np.reshape(X, (len(X), 1))
model = linear_model.LinearRegression()
model.fit(X, np.log(yh))
# calculate trend
trend = model.predict(X)
# plot trend
plt.figure(figsize=(10,8))
plt.plot(range(0,len(yh)),np.log(yh), 'black', label='log(yₜ)')
plt.plot(range(0,len(yh)),trend, 'black', linestyle = "--", label='secular trend')
plt.legend(loc='lower right')
plt.xlabel("Time, quarters")
plt.ylabel("Output per Worker")
plt.show()
```

```python
# detrend
yh_cycle = [np.log(yh[i])-trend[i] for i in range(0, len(yh))]
# plot detrended
plt.figure(figsize=(10,8))
plt.plot(np.exp(yh_cycle), 'black', label='cycle')
#plt.plot(yh_cycle/np.mean(yh_cycle), linestyle = "--", label='cycle normalized')
plt.legend(loc='lower center')
plt.xlabel("time, quarters")
plt.ylabel("yₜ")
plt.show()
```

```python
def solow_swan(t, n, s, δ, α, A, ɛ, λ0, γ, k0):
    """
    A Stochastic, Discrete-Time Solow-Swan Model
    
    Inputs
    ------
      t:            number of periods
      n:            population growth rate
      s:            savings rate
      δ:            depreciation rate
      α:            capital income share 
      A:            technology constant
      ɛ:            stochastic process 
      s:            standard deviation of shock to total factor productivity (TFP)
    
    Outputs
    -------
    - Exact
      ke            capital per income
      ke_tilda:     capital per income, log deviations
      γe:           vdsv
      ye:           dvss
      λe:           csd
    - Approximation
      ke            vds
      ke_tilda:     vsd
      γe:           vdsv
      ye:           dvss
      λe:           csd
    """
        
    
    k_ss = ((s)/(δ+n))**(1/(1-α))
    y_ss = k_ss**α
    
    # Shocks
    λ = np.empty(t); λ[0] = λ0
    for i in range(0,t-1): λ[i+1] = 1+np.sum(np.multiply(np.power(np.repeat(γ,i),range(0,i)), np.flip(ɛ[:i])))+(γ**(i+1))*(λ0-1)
    λ_tilda = np.log(λ)
    
    # exact solution
    ke = np.empty(t); ke[0] = k_ss  
    for i in range(0,t-1): ke[i+1] = ((1-δ)*ke[i]+s*λ[i]*(ke[i]**α))/(1+n)
    
    # approximate solution
    B = (1+α*n-δ*(1-α))/(1+n)
    C = (δ+n)/(1+n)
    ka_tilda = np.empty(t); ka_tilda[0] = np.log(k0)-np.log(k_ss)
    for i in range(0,t-1): ka_tilda[i+1] = C*np.sum(np.multiply(np.power(np.repeat(B, i+1), range(0,i+1)),np.flip(λ_tilda[:i+1])))+(B**(i+1))*ka_tilda[0]
       
    # return
    #ke_tilda = np.log(ke)-np.log(k_ss) 
    #ka = np.exp(ka_tilda+np.log(k_ss))
    #γe = np.log(ke[1:])-np.log(ke[:-1])
    #γa = np.log(ka[1:])-np.log(ka[:-1])
    #ye = A*np.multiply(np.exp(ɛ), np.power(ke,α))
    #ya = A*np.multiply(np.exp(ɛ), np.power(ka,α))
    #ye_tilda = ɛ+np.multiply(α, ke_tilda)
    #ya_tilda = ɛ+np.multiply(α, ka_tilda)
    #λe = np.log(ye[1:])-np.log(ye[:-1])
    #λa = np.log(ya[1:])-np.log(ya[:-1])
    
    return(
        ka_tilda
        #[ke, ke_tilda, γe, ye, ye_tilda, λe],
        #[ka, ka_tilda, γa, ya, ya_tilda, λa],
        #y_ss
    )
```

## Calibration

$$ \text{sd($y_h$)} = 0.0263823561094945,  \quad \text{ log differences} $$ 

```python
sdh = np.std(np.exp(yh_cycle)*y_ss)
```

```python
def g(se): return np.var(solow_swan(t, n, s, δ, α, A, ɛ, se)[0][3]) - sdh**2
```

```python
soi = np.linspace(0, .1, 100)
plt.figure(figsize=(10,8))
plt.plot(soi, soi*0, 'black')
plt.plot(soi, [g(x) for x in soi], 'black', linestyle = "--", label='g(x)')
plt.legend(loc='upper left')
plt.xlabel("σ(ɛ)")
plt.show()
```

```python
sr = brentq(g, 0, 0.1)
```
