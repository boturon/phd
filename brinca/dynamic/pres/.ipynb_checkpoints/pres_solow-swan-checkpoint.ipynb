__log-linear detrending__

$$ y_t \equiv \log{y_t} $$
$$ y_t = a + bt + \varepsilon_t$$
$$ \text{secular trend: } \quad y_t^s = a + bt$$
$$ \text{cycle: } \quad y_t^c = \varepsilon_t$$

```python
X = [i for i in range(0, len(yh))]
X = np.reshape(X, (len(X), 1))
model = linear_model.LinearRegression()
model.fit(X, np.log(yh))
# calculate trend
trend = model.predict(X)
# plot trend
plt.figure(figsize=(10,8))
plt.plot(range(0,len(yh)),np.log(yh), 'black', label='log(y)')
plt.plot(range(0,len(yh)),trend, 'black', linestyle = "--", label='secular trend')
plt.legend(loc='lower right')
plt.xlabel("Time, quarters")
plt.ylabel("Output per Worker")
plt.show()
```

```python
# detrend
yh_cycle = [np.log(yh[i])-trend[i] for i in range(0, len(yh))]
# plot detrended
plt.figure(figsize=(10,8))
plt.plot(np.exp(yh_cycle), 'black', label='cycle')
#plt.plot(yh_cycle/np.mean(yh_cycle), linestyle = "--", label='cycle normalized')
plt.legend(loc='lower center')
plt.xlabel("time, quarters")
plt.ylabel("y")
plt.show()
```

```python
def solow_swan(t, n, s, , , A, , 0, , k0):
    """
    A Stochastic, Discrete-Time Solow-Swan Model
    
    Inputs
    ------
      t:            number of periods
      n:            population growth rate
      s:            savings rate
      :            depreciation rate
      :            capital income share 
      A:            technology constant
      :            stochastic process 
      s:            standard deviation of shock to total factor productivity (TFP)
    
    Outputs
    -------
    - Exact
      ke            capital per income
      ke_tilda:     capital per income, log deviations
      e:           vdsv
      ye:           dvss
      e:           csd
    - Approximation
      ke            vds
      ke_tilda:     vsd
      e:           vdsv
      ye:           dvss
      e:           csd
    """
        
    
    k_ss = ((s)/(+n))**(1/(1-))
    y_ss = k_ss**
    
    # Shocks
     = np.empty(t); [0] = 0
    for i in range(0,t-1): [i+1] = 1+np.sum(np.multiply(np.power(np.repeat(,i),range(0,i)), np.flip([:i])))+(**(i+1))*(0-1)
    _tilda = np.log()
    
    # exact solution
    ke = np.empty(t); ke[0] = k_ss  
    for i in range(0,t-1): ke[i+1] = ((1-)*ke[i]+s*[i]*(ke[i]**))/(1+n)
    
    # approximate solution
    B = (1+*n-*(1-))/(1+n)
    C = (+n)/(1+n)
    ka_tilda = np.empty(t); ka_tilda[0] = np.log(k0)-np.log(k_ss)
    for i in range(0,t-1): ka_tilda[i+1] = C*np.sum(np.multiply(np.power(np.repeat(B, i+1), range(0,i+1)),np.flip(_tilda[:i+1])))+(B**(i+1))*ka_tilda[0]
       
    # return
    #ke_tilda = np.log(ke)-np.log(k_ss) 
    #ka = np.exp(ka_tilda+np.log(k_ss))
    #e = np.log(ke[1:])-np.log(ke[:-1])
    #a = np.log(ka[1:])-np.log(ka[:-1])
    #ye = A*np.multiply(np.exp(), np.power(ke,))
    #ya = A*np.multiply(np.exp(), np.power(ka,))
    #ye_tilda = +np.multiply(, ke_tilda)
    #ya_tilda = +np.multiply(, ka_tilda)
    #e = np.log(ye[1:])-np.log(ye[:-1])
    #a = np.log(ya[1:])-np.log(ya[:-1])
    
    return(
        ka_tilda
        #[ke, ke_tilda, e, ye, ye_tilda, e],
        #[ka, ka_tilda, a, ya, ya_tilda, a],
        #y_ss
    )
```

```python
def g(se): return np.var(solow_swan(t, n, s, , , A, , se)[0][3]) - sdh**2
```

```python
soi = np.linspace(0, .1, 100)
plt.figure(figsize=(10,8))
plt.plot(soi, soi*0, 'black')
plt.plot(soi, [g(x) for x in soi], 'black', linestyle = "--", label='g(x)')
plt.legend(loc='upper left')
plt.xlabel("()")
plt.show()
```

```python
sr = brentq(g, 0, 0.1)
```
